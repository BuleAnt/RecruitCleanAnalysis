## 对搜索到的职业数据进行清洗

```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
from datetime import date
from datetime import timedelta
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import sys
reload(sys)
sys.setdefaultencoding('utf-8')
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签  
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
filePath = "E:\python\DBscrapy\RecruitScrapy\RecruitScrapy\data\getRecruitList.csv"
csvpdfile = pd.read_csv(filePath, encoding='utf-8')
#csvpdfile.head(200)
#csvpdfile.columns
```

```python
#移除空行超过一半的数据行
csvClear1 = csvpdfile.dropna(axis=0,thresh=int(len(csvpdfile.columns)/4))
#csvClear1 = csvpdfile.dropna()
```

```python
#对列中的空值进行操作
#替换空置,并清除多余的值
csvClear1['respond'] = csvClear1['respond'].replace(np.nan,'20%')
csvClear1['companySize'] = csvClear1['companySize'].str.extract(u'([\d]+-[\d]+|[\d]+)')
csvClear1['experience'] = csvClear1['experience'].str.replace(u'不限','0-1')
csvClear1['experience'] = csvClear1['experience'].str.extract(u'([\d]+-[\d]+)')
csvClear1['education'] = csvClear1['education'].str.extract(u'：([\u4e00-\u9fa5]+)')
csvClear1['companyType'] = csvClear1['companyType'].str.extract(u'：([\u4e00-\u9fa5]+)')
csvClear1['companyType'] = csvClear1['companyType'].replace(np.nan,'民营')
csvClear1['companyPlace'] = csvClear1['companyPlace'].str.extract(u'：([\u4e00-\u9fa5]+)')


#替换时间.今天，昨天，前天
newDate = date.today()
today = newDate.strftime('%m-%d')
yesterday = (newDate-timedelta(days=1)).strftime('%m-%d')
beforeday = (newDate-timedelta(days=2)).strftime('%m-%d')
day15 = (newDate-timedelta(days=15)).strftime('%m-%d')


csvClear1['releaseDate'] = csvClear1['releaseDate'].replace([u'今天',u'昨天',u'前天','15天前'],[today,yesterday,beforeday,day15])
csvClear1['releaseDate'] = csvClear1['releaseDate'].str.extract('([\d]+-[\d]+)')
csvClear1['releaseDate'] = csvClear1['releaseDate'].replace(np.nan,today)


DatfFile = "E:\python\DBscrapy\RecruitScrapy\RecruitScrapy\data\getRecruitList1.csv"
csvClear1.to_csv(DatfFile, encoding='utf-8', mode='a')#导出数据到CSV文件

csvClear1.head()

```

```python
#获得地区公司统计
placeData = []
for name,group in csvClear1.groupby(['companyPlace','companyName']):
    #placeData.append([name[0].encode("utf-8"),name[1].encode("utf-8")]) 
    placeData.append([name[0],name[1]]) 

placeDF = pd.DataFrame(placeData, columns=['companyPlace','companyName']) 


placeCompanyValue = {}
for name,group in placeDF.groupby('companyPlace'):
    placeCompanyValue[name] = len(group.index)
    print  name
```

```python
#读取坐标信息
countries = {}
endataPath = "E:\python\DBscrapy\RecruitScrapy\RecruitScrapy\data\chineseENdata.csv"
enfile = pd.read_csv(endataPath, encoding='utf-8')
enfile.head()

for inte in enfile.values:
    countries[inte[2]]=[inte[3],inte[4],20.0]
```

```python
#绘制地图

# 根据城市公司值更新点的大小
for i in placeCompanyValue.keys():
    countries[i][2] = 0.1*placeCompanyValue[i]
plt.figure(figsize=(24,24))

#   llcrnrlat - 左下角的纬度   北纬N20°28′5.48″ 东经E85°25′46.88″
#   llcrnrlon - 左下角的经度
#   urcrnrlat - 右上角的纬度   北纬N49°05′35.56″ 东经E129°43′6.00″
#   urcrnrlon - 右上角的经度
mp = Basemap(projection='cyl',    # 地图投影方式 
                  llcrnrlon=90.00, llcrnrlat=21.00, urcrnrlon=132.00,urcrnrlat=45.00, # 经纬度范围
                   resolution='l')

mp.drawcountries(linewidth=1) #绘制出国家
mp.fillcontinents(color='lavender', lake_color='#907099')
mp.drawmapboundary(linewidth=0.2, fill_color='#4169E1')
# mp.drawparallels(np.arange(-90,90,30),labels=[0,0,0,0], color='white', linewidth=0.5)
# mp.drawmeridians(np.arange(0,360,30),labels=[0,0,0,0], color='white', linewidth=0.5)
    
# 绘制城市
for i in placeCompanyValue.keys():
        mp.plot(countries[i][0], countries[i][1], 'bo', markersize = countries[i][2], color='r')

# 添加城市名称
for key in placeCompanyValue.keys():
    if countries[i][2]>1:
        plt.text(countries[key][0] - 0.85, countries[key][1] , key, fontsize = 14, color="black")

plt.title('数据公司集中点')
plt.show()

# 保存数据的可视化
plt.savefig('./country_rank.png')
```

```python
#使用堆积条状图 生成一个适合投递公司列表

csvClear2 = csvClear1.copy()

#生成公司规模列表
scaleList = ['20','20-99','100-499','500-999','1000-9999','10000','0-1','1-3','3-5','5-10']
scaleValueList = [4,8,10,12,14,18,4,4,4,4]
csvClear2['companySize'] = csvClear2['companySize'].replace(scaleList,scaleValueList)
csvClear2['companySize'] = csvClear2['companySize'].replace(np.nan,4)
csvClear2['companySize'] = csvClear2['companySize'].astype('float16')
csvClear2['companySize'] = csvClear2['companySize'].astype('int')

#对工资求中位数，变成k
csvClear2['wagesMin'] = csvClear2['wages'].str.extract('([\d]+)-')
csvClear2['wagesMax'] = csvClear2['wages'].str.extract('-([\d]+)')
csvClear2['wagesMin'] = csvClear2['wagesMin'].astype('float16')
csvClear2['wagesMax'] = csvClear2['wagesMax'].astype('float16')

csvClear2['wages'] = csvClear2['wagesMin']+csvClear2['wagesMax']
csvClear2['wages'] = csvClear2['wages']/2000
#csvClear2['wages'] = csvClear2['wagesMax'].astype('int16')

csvClear2['wages'] = csvClear2['wages'].replace(np.nan,3)

#experience,生成工作经验列表
experienceList = ['0-1','1-3','3-5','5-10']
scaleValueList = [8,6,4,1]
csvClear2['experience'] = csvClear2['experience'].replace(experienceList,scaleValueList)

#生成反馈率
csvClear2['respond'] = csvClear2['respond'].str.extract('([\d])')
csvClear2['respond'] = csvClear2['respond'].replace(np.nan,2)
csvClear2['respond'] = csvClear2['respond'].astype('int')


```
